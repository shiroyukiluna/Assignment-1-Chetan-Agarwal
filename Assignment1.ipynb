## 4) Z-test score

```python
import pandas as pd
from statsmodels.stats.weightstats import ztest

df = pd.read_csv(r"placement_data_01.csv")

# Selecting relevant numerical columns for the test
numerical_cols = ['Internships','Projects','Workshops/Certifications']

# Performing pairwise Z-tests
z_test_results = {}

for i in range(len(numerical_cols)):
    for j in range(i + 1, len(numerical_cols)):
        col1, col2 = numerical_cols[i], numerical_cols[j]
        z_stat, p_value = ztest(df[col1].dropna(), df[col2].dropna())
        z_test_results[f"{col1} vs {col2}"] = (z_stat, p_value)

# Display results
for pair, (z_stat, p_value) in z_test_results.items():
    print(f"{pair}: z-statistic = {z_stat:.4f}, p-value = {p_value:.4f}")

```

#### Summary
##### The Z-test results highlight significant differences between internships, projects, and workshops/certifications. The highly negative z-statistic for "Internships vs Projects" suggests that internship counts are significantly lower than project counts. Similarly, the large positive z-statistic for "Projects vs Workshops/Certifications" indicates that projects are significantly more prevalent than workshops/certifications. The comparison between "Internships vs Workshops/Certifications" has a smaller z-statistic but still a statistically significant p-value (0.0013), suggesting a moderate difference. Overall, these results imply that students tend to undertake more projects compared to internships and certifications, which could influence their skill development and placement outcomes.
